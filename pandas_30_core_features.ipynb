# Pandas 30 个基础核心功能演示

本 notebook 通过两个精简的虚构数据集，系统演示 pandas 最常用的 30 个核心功能，
覆盖**数据查看、选择、筛选、清洗、转换、聚合、合并、透视**等日常分析全流程。

- **数据集 1 `employees`**：员工信息表（含部门、年龄、薪资、入职日期等，故意留了缺失值）
- **数据集 2 `sales`**：销售订单表（可与员工表通过 `emp_id` 关联）

> 运行方式：从上到下依次执行（Shift+Enter）。每个小节标题即对应的 pandas 功能点。

## 0. 准备：导入库 & 构造虚构数据集

```python
import pandas as pd
import numpy as np

print("pandas 版本:", pd.__version__)
```

```python
# 数据集 1：员工信息表（department / salary 故意留有缺失值 NaN）
employees = pd.DataFrame({
    "emp_id":     [1, 2, 3, 4, 5, 6, 7, 8],
    "name":       ["张伟", "李娜", "王芳", "刘洋", "陈静", "杨磊", "赵敏", "周强"],
    "department": ["研发", "市场", "研发", "销售", "市场", "研发", "销售", np.nan],
    "age":        [29, 34, 41, 28, 45, 31, 38, 26],
    "salary":     [21000, 18000, 27000, 15000, 23000, np.nan, 16000, 14000],
    "hire_date":  ["2019-03-01", "2017-07-15", "2015-01-20", "2020-09-10",
                   "2014-05-30", "2018-11-05", "2016-08-22", "2021-02-14"],
    "city":       ["北京", "上海", "北京", "深圳", "上海", "杭州", "深圳", "北京"],
    "gender":     ["M", "F", "F", "M", "F", "M", "F", "M"],
})

employees
```

```python
# 数据集 2：销售订单表（通过 emp_id 可与员工表关联）
sales = pd.DataFrame({
    "order_id":   range(1001, 1013),
    "emp_id":     [1, 2, 1, 4, 7, 4, 2, 5, 1, 7, 4, 3],
    "product":    ["A", "B", "C", "A", "B", "A", "C", "B", "A", "C", "B", "A"],
    "quantity":   [3, 5, 2, 7, 1, 4, 6, 2, 9, 3, 5, 1],
    "price":      [199, 99, 299, 199, 99, 199, 299, 99, 199, 299, 99, 199],
    "order_date": ["2023-01-05", "2023-01-08", "2023-02-11", "2023-02-15",
                   "2023-03-01", "2023-03-09", "2023-03-20", "2023-04-02",
                   "2023-04-18", "2023-05-05", "2023-05-21", "2023-06-01"],
})

sales
```

## 1. `head()` / `tail()` / `sample()` — 快速预览数据

查看数据的前几行、后几行或随机抽样，是拿到数据的第一步。

```python
print("前 3 行：")
display(employees.head(3))

print("后 2 行：")
display(employees.tail(2))

print("随机抽样 3 行：")
employees.sample(3, random_state=42)
```

## 2. 基本属性 — `shape` / `columns` / `dtypes` / `index`

不需要括号，它们是 DataFrame 的**属性**，用来了解数据的"骨架"。

```python
print("形状 (行, 列):", employees.shape)
print("列名:", list(employees.columns))
print("行数:", len(employees))
print("\n各列数据类型:")
print(employees.dtypes)
```

## 3. `info()` — 结构概览（类型 + 非空计数 + 内存）

一眼看出哪些列有缺失值、各列类型、整体占用内存。

```python
employees.info()
```

## 4. `describe()` — 描述性统计

默认对数值列给出计数、均值、标准差、分位数；加 `include="all"` 可一并描述文本列。

```python
print("数值列描述统计：")
display(employees.describe())

print("全部列（含文本列）：")
employees.describe(include="all")
```

## 5. 选择列 — 单列返回 Series，多列返回 DataFrame

```python
# 单列 -> Series
s = employees["name"]
print(type(s).__name__, "→\n", s.head(3), sep="")

# 多列 -> DataFrame（注意双层方括号）
employees[["name", "department", "salary"]].head()
```

## 6. `loc` — 按**标签**选择行和列

语法：`df.loc[行标签, 列标签]`，区间是**闭区间**（含末端）。

```python
# 第 0~2 行的 name、salary 列
display(employees.loc[0:2, ["name", "salary"]])

# 配合条件：研发部门的姓名与薪资
employees.loc[employees["department"] == "研发", ["name", "salary"]]
```

## 7. `iloc` — 按**位置**（整数下标）选择

语法：`df.iloc[行位置, 列位置]`，区间是**左闭右开**（同 Python 切片）。

```python
# 前 3 行、前 4 列
display(employees.iloc[0:3, 0:4])

# 取单个值：第 2 行第 1 列
print("employees.iloc[2, 1] =", employees.iloc[2, 1])
```

## 8. 布尔条件筛选 — 最常用的"按条件取数据"

```python
# 薪资 > 18000 的员工
high_pay = employees[employees["salary"] > 18000]
high_pay[["name", "salary"]]
```

## 9. 多条件 & `isin()` / `between()`

多条件用 `&`（与）`|`（或），每个条件必须用括号包起来。

```python
# 年龄在 30~40 之间，且不在北京
mask = employees["age"].between(30, 40) & (employees["city"] != "北京")
display(employees[mask][["name", "age", "city"]])

# isin：城市属于一线城市集合
employees[employees["city"].isin(["北京", "上海"])][["name", "city"]]
```

## 10. 新增列 — 直接赋值 / `assign()`

```python
# 直接赋值新增一列：年薪
employees["annual_salary"] = employees["salary"] * 12

# assign 链式新增（返回新对象，不改原表）：是否高级员工
employees = employees.assign(senior=lambda d: d["age"] >= 40)

employees[["name", "salary", "annual_salary", "senior"]]
```

## 11. `drop()` — 删除列或行

`axis=1` 删列，`axis=0`（默认）删行。

```python
# 删除刚才的辅助列（不改原表，返回副本）
tmp = employees.drop(columns=["annual_salary", "senior"])
print("删列后剩余列：", list(tmp.columns))

# 删除指定行（按索引标签）
print("删行后行数：", employees.drop(index=[0, 1]).shape[0])
```

## 12. `rename()` — 重命名列名 / 索引

```python
employees.rename(columns={
    "name": "姓名",
    "department": "部门",
    "salary": "月薪",
}).head(3)
```

## 13. `sort_values()` — 按值排序

```python
# 按薪资降序；缺失值放最后
employees.sort_values("salary", ascending=False, na_position="last")[["name", "salary"]]
```

## 14. `set_index()` / `reset_index()` — 设置与重置索引

```python
# 将 emp_id 设为索引
indexed = employees.set_index("emp_id")
display(indexed.loc[[3, 5], ["name", "salary"]])   # 用 emp_id 直接取数

# 还原为默认整数索引
indexed.reset_index().head(3)
```

## 15. 缺失值检测 — `isnull()` / `notnull()` / `.sum()`

```python
# 每列缺失值数量
print("各列缺失值数量：")
print(employees.isnull().sum())

# 含缺失值的行
print("\n含缺失值的行：")
employees[employees.isnull().any(axis=1)][["name", "department", "salary"]]
```

## 16. `fillna()` — 填充缺失值

```python
filled = employees.copy()
# 数值列用中位数填充，类别列用众数/固定值填充
filled["salary"] = filled["salary"].fillna(filled["salary"].median())
filled["department"] = filled["department"].fillna("未分配")

filled[["name", "department", "salary"]]
```

## 17. `dropna()` — 删除缺失值

```python
# 删除任何含缺失值的行
print("dropna 后行数：", employees.dropna().shape[0])

# 只在 salary 列缺失时删除
employees.dropna(subset=["salary"]).shape[0]
```

## 18. `astype()` — 类型转换

```python
conv = employees.dropna(subset=["salary"]).copy()
# 浮点薪资转为整数；gender 转为 category 类型（更省内存）
conv["salary"] = conv["salary"].astype(int)
conv["gender"] = conv["gender"].astype("category")

print(conv[["salary", "gender"]].dtypes)
conv[["name", "salary", "gender"]].head(3)
```

## 19. `to_datetime()` 与 `.dt` 访问器 — 日期处理

```python
employees["hire_date"] = pd.to_datetime(employees["hire_date"])

# 用 .dt 提取年份，并计算工龄
employees["hire_year"] = employees["hire_date"].dt.year
employees["tenure_years"] = 2023 - employees["hire_year"]

employees[["name", "hire_date", "hire_year", "tenure_years"]]
```

## 20. `value_counts()` — 频数统计

```python
# 各城市的员工数（normalize=True 可得到占比）
print("各城市人数：")
print(employees["city"].value_counts())

print("\n各城市占比：")
employees["city"].value_counts(normalize=True).round(3)
```

## 21. `unique()` / `nunique()` — 去重值与去重计数

```python
print("出现过的部门:", employees["department"].unique())
print("不同部门数量 (不含 NaN):", employees["department"].nunique())
print("不同城市数量:", employees["city"].nunique())
```

## 22. `groupby()` — 分组聚合（最核心的分析操作）

```python
# 各部门的平均薪资与人数
employees.groupby("department")["salary"].mean().round(0)
```

```python
# 多列分组：城市 × 性别 的平均年龄
employees.groupby(["city", "gender"])["age"].mean().round(1)
```

## 23. `agg()` — 一次做多种聚合

```python
# 各部门：薪资的均值/最大/最小 + 人数
employees.groupby("department").agg(
    人数=("emp_id", "count"),
    平均薪资=("salary", "mean"),
    最高薪资=("salary", "max"),
    平均年龄=("age", "mean"),
).round(1)
```

## 24. `apply()` — 应用自定义函数

`apply` 可作用于 Series（逐元素/逐组）或 DataFrame（逐行/逐列）。

```python
# 对 Series：把薪资分档
def pay_level(x):
    if pd.isna(x):
        return "未知"
    if x >= 22000:
        return "高"
    if x >= 16000:
        return "中"
    return "低"

employees["pay_level"] = employees["salary"].apply(pay_level)

# 对 DataFrame 逐行：拼接展示字符串
employees["label"] = employees.apply(
    lambda row: f"{row['name']}({row['city']})", axis=1
)

employees[["name", "salary", "pay_level", "label"]]
```

## 25. `map()` / `replace()` — 值映射与替换

```python
# map：用字典把 M/F 映射为中文
employees["gender_cn"] = employees["gender"].map({"M": "男", "F": "女"})

# replace：替换部门名称
employees["department"] = employees["department"].replace({"研发": "研发部", "市场": "市场部"})

employees[["name", "gender", "gender_cn", "department"]]
```

## 26. `.str` 字符串访问器 — 文本列批量操作

```python
# 标签列转大写、判断是否包含某城市、取长度
demo = sales.copy()
print("product 列加前缀：")
print(("品类-" + sales["product"]).head(3).tolist())

print("\n员工姓名长度：")
print(employees["name"].str.len().tolist())

print("\n姓名是否姓'王':")
print(employees["name"].str.startswith("王").tolist())
```

## 27. `merge()` — 表连接（类似 SQL JOIN）

把销售表与员工表按 `emp_id` 关联，给每笔订单补上员工信息。

```python
merged = sales.merge(
    employees[["emp_id", "name", "department", "city"]],
    on="emp_id",
    how="left",          # 左连接：保留所有订单
)
merged.head()
```

## 28. `concat()` — 纵向 / 横向拼接

```python
# 纵向拼接：把两个月份的子表上下叠加
jan = sales[sales["order_date"].str.startswith("2023-01")]
feb = sales[sales["order_date"].str.startswith("2023-02")]

combined = pd.concat([jan, feb], ignore_index=True)
print("1 月 + 2 月订单数:", len(combined))
combined
```

## 29. `pivot_table()` — 透视表

行=员工、列=产品，值=销售额合计，快速得到二维汇总。

```python
sales["amount"] = sales["quantity"] * sales["price"]

pivot = sales.pivot_table(
    index="emp_id",
    columns="product",
    values="amount",
    aggfunc="sum",
    fill_value=0,
    margins=True,          # 加总计行/列
)
pivot
```

## 30. `cut()` / `qcut()` — 数值分箱（连续变量离散化）

- `cut`：按**自定义区间**分箱
- `qcut`：按**分位数**等频分箱

```python
# cut：按指定年龄段分组
employees["age_group"] = pd.cut(
    employees["age"],
    bins=[20, 30, 40, 50],
    labels=["20-30", "30-40", "40-50"],
)

# qcut：按薪资三等分（低/中/高）
employees["salary_tier"] = pd.qcut(
    employees["salary"], q=3, labels=["低", "中", "高"]
)

employees[["name", "age", "age_group", "salary", "salary_tier"]]
```

## 附加：4 个高频补充功能

虽然主清单是 30 个，这里再补充几个日常分析极常用的小工具。

### A1. `duplicated()` / `drop_duplicates()` — 去重

```python
dup_demo = pd.DataFrame({"city": ["北京", "上海", "北京", "深圳", "上海"]})
print("是否重复:", dup_demo["city"].duplicated().tolist())
print("去重后:", dup_demo["city"].drop_duplicates().tolist())
```

### A2. `crosstab()` — 交叉频数表

```python
# 城市 × 性别 的人数交叉表
pd.crosstab(employees["city"], employees["gender"], margins=True)
```

### A3. 累计 / 滚动 — `cumsum()` / `rolling()`

```python
ts = sales.sort_values("order_date").copy()
ts["amount"] = ts["quantity"] * ts["price"]
ts["cum_amount"] = ts["amount"].cumsum()                 # 累计销售额
ts["rolling3"] = ts["amount"].rolling(3).mean().round(1) # 3 期移动平均
ts[["order_date", "amount", "cum_amount", "rolling3"]]
```

### A4. `to_csv()` / `read_csv()` — 读写文件

```python
# 写出再读回，验证往返一致
employees.to_csv("employees_demo.csv", index=False, encoding="utf-8-sig")
reloaded = pd.read_csv("employees_demo.csv")
print("写出并读回成功，形状:", reloaded.shape)
reloaded.head(2)
```
